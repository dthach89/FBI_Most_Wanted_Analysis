```python
import pandas as pd
from datetime import datetime as dt
```

```python
bank_df = pd.read_csv('Paul_visualization/bank-full_1.csv')
bank_df
```

```python
bank_df['month'].describe
```

```python
look_up = {'january': '01' , 'febuary': '02' , 'march': '03' , 'april': '04' , 'may': '05',
            'june': '06', 'july': '07' , 'august': '08' , 'september': '09' , 'october': '10' , 'november': '11' , 'december': '12' }

bank_df['month'] = bank_df['month'].apply(lambda x: look_up[x])
bank_df
```

```python
bank_df['date'] = pd.to_datetime(bank_df[['day', 'month', 'year']])
bank_df
```

```python
bank_df = bank_df.drop(['day', 'month', 'year'], axis=1)
bank_df
```

```python
bank_df
```

```python
bank_df['weekday'] = bank_df['date'].dt.day_name()
bank_df
```

```python
bank_df = bank_df.to_csv('Paul_visualization/bank-full_2.csv')
```
